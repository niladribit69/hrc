#Renaming Columns using rename method

```python
df.rename(columns={'City':'city','State':'country'},inplace=True)
df.head()
```

# Deleting Rows and Columns in pandas dataframe

#Removing single column using drop method

```python
df.drop('Colours reported',axis=1,inplace=True)
df.head()
```

#Removing multiple columns using drop method

```python
df.drop(['City','State'],axis=1,inplace=True)
df.head()
```

#Removing rows from dataframe using drop method 

```python
df.drop([0,1],axis=0,inplace=True)
df.head()
```

```python
grouped=df.groupby('genre')

for name,group in grouped:
    print(name)
    print(group)
```
